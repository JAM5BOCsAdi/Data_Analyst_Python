```python
import requests
from bs4 import BeautifulSoup

my_headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

def scrape_arukereso_pages(base_url):
    all_data = []
    page_num = 1
    
    while True:
        url = f"{base_url}?start={(page_num - 1) * 25}"
        response = requests.get(url, headers=my_headers)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            data = extract_data_from_page(soup)
            
            if not data:
                break  # No more data on the page
            
            all_data.extend(data)
            if len(data) < 25:
                print_items(all_data, page_num)
                break
            
            if len(all_data) >= 25:
                print_items(all_data[:25], page_num)
                all_data = all_data[25:]
                page_num += 1
            else:
                page_num += 1
        else:
            print(f"Failed to fetch page {page_num}")
            break

def extract_data_from_page(soup):
    titles = soup.find_all('div', class_='name ulined-link')
    data = [title.text.strip() for title in titles]
    return data

def print_items(items, page_num):
    print(f"---------------- Page {page_num} ----------------")
    for i, item in enumerate(items, start=1):
        print(f"{i}. {item}")

# Example usage
base_url = "https://www.arukereso.hu/notebook-c3100/"
scrape_arukereso_pages(base_url)
```
